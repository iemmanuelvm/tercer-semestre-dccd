```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties


sns.set_style("white")
plt.rc('font', family='Times New Roman')

df1 = pd.read_csv('_training_metrics_stft.csv')
df2 = pd.read_csv('training_metrics_stft.csv')

fig, ax = plt.subplots(figsize=(22, 14))

ax.plot(df1.index, df1['Train Source Acc'], label='Training Accuracy (Source)', linewidth=1, linestyle='-', marker='x', markersize=8)
ax.plot(df1.index, df1['Train Target Acc'], label='Training Accuracy (Target)', linewidth=1, linestyle='--', marker='s', markersize=8)
ax.plot(df1.index, df1['Val Source Acc'], label='Validation Accuracy (Source)', linewidth=1, linestyle='-.', marker='^', markersize=8)
ax.plot(df1.index, df1['Val Target Acc'], label='Validation Accuracy (Target)', linewidth=1, linestyle=':', marker='d', markersize=8)
ax.plot(df2.index, df2['Train Target Acc'], label='Training Accuracy\n(Model without Domain Adaptation)', linewidth=2, linestyle='--', marker='o', markersize=8)
ax.plot(df2.index, df2['Val Target Acc'], label='Validation Accuracy\n(Model without Domain Adaptation)', linewidth=2, linestyle=':', marker='v', markersize=8)

ax.set_xlabel('Epochs', fontsize=90, fontweight='bold')
ax.set_ylabel('Accuracy', fontsize=90, fontweight='bold')
ax.set_ylim(0.5, 1)
ax.set_xticks(range(0, max(len(df1.index), len(df2.index)) + 10, 10))
ax.tick_params(axis='x', labelsize=60)
ax.tick_params(axis='y', labelsize=60)
ax.grid(True, linestyle=':', linewidth=1)
ax.set_facecolor('white')

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels,
           loc='lower center',
           bbox_to_anchor=(0.5, -0.24), 
           ncol=2,
           prop=FontProperties(weight='bold', size=45),
           frameon=True, fancybox=True, framealpha=1)

plt.subplots_adjust(bottom=0.28)

plt.show()
```
